```python
"""
Looks like list comprehension, but with round brackets (). 
- It returns a generator (lazy iterator), not a list.
- Used with large data sets, when we dont know how many to loop through.
"""
gen = (x*x for x in range(3))
print(gen)        # Generator object
print(next(gen))  # 0
print(next(gen))  # 1
```

```python
def returnsquares(n):
    for i in range(1, 11):
        yield i**2

print(next(returnsquares(4)))
```

```python
"""
An iterator is an object that can be traversed one element at a time using the built-in functions:
- iter() → to get an iterator from an iterable
- next() → to fetch the next element
It remembers its current state, so you can resume where you left off.
"""
numbers = [1,2,3,4,5]
iterator = iter(numbers)

# Prints firs 3 elements as we gave given next() 3 times below.
print(next(iterator))
print(next(iterator))
print(next(iterator))

```

```python
"""
A generator is a special kind of iterator that:
- Is defined using yield instead of return
- Is lazy — generates values one at a time on demand, saving memory
"""
def read_file_line_by_line(filename):
    with open(filename) as f:
        for line in f:
            yield line.strip()
```
