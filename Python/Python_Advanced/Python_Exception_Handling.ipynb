```python
try:                    # Put everything in try where we feel like some exception can occur
    x = 5 + 3
except:                 # Handles the error using except
    print("Error")
else:                   # Only, Executes if no exception occurred.
    print("Success:", x)
finally:                # Always executed
    print("Closing operation or logging.")
```

```python
age = -5
if age < 0:
    raise Exception("Age cannot be negative.")
```

```python
try:
    with open("data.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("The file does not exist.")
```

```python
class MyCustomError(Exception):
    pass

def do_something(x):
    if x < 0:
        raise MyCustomError("x cannot be negative")

try:
    do_something(-1)
except MyCustomError as e:
    print("Caught:", e)
```
