```python
f = open('../Test_Files/myfile.txt', 'w')  # Open in write mode
f.write("this is testing writing to a file")

f.close()   #closing a file
```

```python
f = open('../Test_Files/myfile.txt', 'r')  # Open in write mode
print(f.read())

f.close()   #closing a file
```

```python
| Mode  | Meaning                           |
| ----- | --------------------------------- |
| `'r'` | Read (default), file must exist   |
| `'w'` | Write (truncates if file exists)  |
| `'a'` | Append to file                    |
| `'x'` | Create, fail if file exists       |
| `'b'` | Binary mode (e.g., `'rb'`)        |
| `'t'` | Text mode (default, e.g., `'rt'`) |
| `'+'` | Read and Write (e.g., `'r+'`)     |
```

```python
with open("../Test_Files/with_open.txt","wb") as f:
    f.write(b"This is test using with open()")
```

```python
with open("../Test_Files/with_open.txt","rb") as f:
    print(f.read())
```

```python
with open("../Test_Files/with_open.txt", "r") as f:
    for lines in f:
        print(lines.strip())
```
