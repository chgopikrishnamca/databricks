```python
import mysql.connector

# Connect to MySQL server
conn = mysql.connector.connect(
    host='localhost',
    user='your_username',
    password='your_password',
    database='test_db'  # You can skip this initially if DB doesn't exist
)
cursor = conn.cursor()

# --- CREATE DATABASE and TABLE ---
cursor.execute("CREATE DATABASE IF NOT EXISTS test_db")
cursor.execute("USE test_db")
cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    id INT PRIMARY KEY AUTO_INCREMENT,
    name VARCHAR(100),
    role VARCHAR(50),
    salary DECIMAL(10, 2)
)
""")

# --- C: CREATE (Insert data) ---
cursor.execute("INSERT INTO employees (name, role, salary) VALUES (%s, %s, %s)",
               ('Alice', 'Engineer', 75000))
cursor.execute("INSERT INTO employees (name, role, salary) VALUES (%s, %s, %s)",
               ('Bob', 'Manager', 90000))
conn.commit()

# --- R: READ (Query data) ---
cursor.execute("SELECT * FROM employees")
rows = cursor.fetchall()
print("\n--- Employees ---")
for row in rows:
    print(row)

# --- U: UPDATE ---
cursor.execute("UPDATE employees SET salary = %s WHERE name = %s", (80000, 'Alice'))
conn.commit()

# --- D: DELETE ---
cursor.execute("DELETE FROM employees WHERE name = %s", ('Bob',))
conn.commit()

# Final read to verify update and delete
cursor.execute("SELECT * FROM employees")
print("\n--- After Update/Delete ---")
for row in cursor.fetchall():
    print(row)

# --- Clean up ---
cursor.execute("DROP TABLE employees")
cursor.execute("DROP DATABASE test_db")

# Close connection
cursor.close()
conn.close()
```
