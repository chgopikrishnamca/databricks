```python
"""
Counter: 
Counts the occurrences of elements in a list/tuple/strings using Counter in Collections Framework
Using Counter in Set referes to give always 1 in o/p as set doesnt allows duplicates.
We can use Counter with dictionary as well when using dict.keys() or dict.values()

Exampe:
Count the occurrences of elements in a list using Counter in Collections Framework
Input: ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
Expected Output: {'apple': 3, 'banana': 2, 'orange': 1}
"""
from collections import Counter
list = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

print(Counter(list))
```

```python
"""
DefaultDict: 
Assigns default values to keys in a dictionary.

Example:
Count the occurrences of elements in a list using Default Dictionary in Collections Framework
Input: ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
Expected Output: {'apple': 3, 'banana': 2, 'orange': 1}
"""
from collections import defaultdict
list = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
dict = defaultdict(int)

for item in list:
    dict[item] +=1

print(dict)
```

```python
"""
Ordered Dict: 
Preserves the insertion order of elements in a dictionary.
NOTE: Python mainatains insertion order for dictionaries om or after python 3.7 versions
Example to use Ordered dictionary in Collections Framework.
"""
from collections import OrderedDict
general_dict={}
general_dict['week1']="Monday"
general_dict['week2']="Tuesday"
general_dict['week3']="Wednesday"
general_dict['week4']="Thursday"
general_dict['week5']="Friday"
general_dict['week6']="Saturday"
general_dict['week7']="Sunday"

print(f"General Dictionary {general_dict}")

ordered_dict=OrderedDict()
ordered_dict['week1']="Monday"
ordered_dict['week2']="Tuesday"
ordered_dict['week3']="Wednesday"
ordered_dict['week4']="Thursday"
ordered_dict['week5']="Friday"
ordered_dict['week6']="Saturday"
ordered_dict['week7']="Sunday"

print(f"Ordered Dictionary {ordered_dict}")
```

```python
"""
Deque :
deque stands for Double-Ended Queue, meaning:
- You can append and pop elements from both ends (front and rear).
- It takes a lsit as an argument and returns a deque object
- It’s faster than a list for insertions and deletions at either end (because it’s implemented as a doubly linked list)
"""
from collections import deque

# Create a deque
dq = deque([1, 2, 3])

# Append elements
dq.append(4)                                            # Adds to right
print(f"Append element 4 on the right side: {dq}")
dq.appendleft(6)       
print(f"Append element 6 on the left side: {dq}")   


# Pop elements
dq.pop()                                                # Removes from right (4)
print(f"Remove element 4 on the right side: {dq}")           
dq.popleft()                                            # Removes from left (6)
print(f"Remove element 6 on the left side: {dq}")        

print(f"Final list: {dq}")           # deque([1, 2, 3])
```

```python
"""
ChainMap: 
ChainMap is useful to link multiple dictionaries together and create a single view of them.
It is useful when you want to combine multiple dictionaries into a single view.
"""

from collections import ChainMap 
   
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
d3 = {'e': 5, 'f': 6}

# Defining the chainmap 
c = ChainMap(d1, d2, d3) 
print(c)
```
