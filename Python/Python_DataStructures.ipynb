```python
"""
Flatten a nested list of unknown depth
Input: [1, [2, [3, 4], 5], 6]
Output: [1, 2, 3, 4, 5, 6]
"""
nested_list = [1, [2, [3, 4], 5], 6]

def flatten_list(nested_list):
    flat_list = []

    for item in nested_list:
        if isinstance(item, list):
            flat_list.extend(flatten_list(item))
        else:
            flat_list.append(item)
    
    return flat_list

print(flatten_list(nested_list))
```

```python
lst = [1, 2, 2, 3, 1, 4]
# show duplicates

duplicate_list = set()
def duplicates(lst):    
    for i in lst:
        if lst.count(i)>1:
            duplicate_list.add(i)
    
    return duplicate_list

values=duplicates(lst)
print(values)
```

```python
"""
Remove duplicates from the list while maintaining order
Input: [1, 2, 2, 3, 1, 4]
Output: [1, 2, 3, 4]
"""

list = [1, 2, 2, 3, 1, 4]
unique_list = []
def remove_duplicates(list):
    for item in list:
        if item not in unique_list:
            unique_list.append(item)

remove_duplicates(list)

print(f"Given List => {list}")
print(f"Unique List => {unique_list}")
```

```python
"""
Count the occurrences of elements in a list
Input: ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
Expected Output: {'apple': 3, 'banana': 2, 'orange': 1}
"""
list = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
count_words = {}        # dictionary to store items and their counts
def count_occurance(list):
    for item in list:
        if item in count_words:
            count_words[item] += 1          # key's value is incremented by 1
        else:
            count_words[item] = 1           # key's value remains 1, as no other match is found

count_occurance(list)
print(count_words)
```

```python
"""
Reverse a string without using built-in reverse() or slicing
Input: "hello"
Output: "olleh
"""
input = "hello"
reverse_slice = input[::-1]    # Using slicing

# Using general function
def reverse_string(input):   
    reverse_word = "" 
    for char in input:
        reverse_word = char + reverse_word
    return reverse_word

reverse = reverse_string(input)

print(f"Reverse string using slicing {reverse_slice}")
print(f"Reverse string using reverse function: {reverse}")
```

```python
"""
Group anagrams from a list of words
Input: ["bat", "tab", "tap", "pat", "cat"]
Output: [["bat", "tab"], ["tap", "pat"], ["cat"]]
"""
words = ["bat", "tab", "tap", "pat"]
anagrams = []

def group_anagrams(words):
    for word in words:
        if word in words and word[::-1] in words:
            anagrams.append([word, word[::-1]])

group_anagrams(words)
print(anagrams)
```

```python
"""
Swap keys and values in a dictionary
Input: {'a': 1, 'b': 2, 'c': 3}
Output: {1: 'a', 2: 'b', 3: 'c'}
"""
input_dict = {'a': 1, 'b': 2, 'c': 3}
output_dict = {}

for key, value in input_dict.items():
    output_dict[value] = key

print(output_dict)
print({k:v for v,k in input_dict.items()}) # Using dictionary comprehensions
```

```python
"""
Merge two dictionaries. If keys overlap, sum their values
Input: d1 = {'a': 5, 'b': 3}, d2 = {'a': 2, 'c': 8}
Output: {'a': 7, 'b': 3, 'c': 8}
"""
from collections import Counter

d1 = {'a': 5, 'b': 3}
d2 = {'a': 2, 'c': 8}

merged_dict = dict(Counter(d1) + Counter(d2))
print(merged_dict)  # {'a': 7, 'b': 3, 'c': 8}
```

```python
"""
Write a function to check if a string is a palindrome (ignore case and spaces)
Input: "A man a plan a canal Panama"
Output: True
"""
input_string = "A man a plan a canal Panama"
input_list = input_string.split()
input_list_str="".join(input_list)
input_lower_string = input_list_str.lower()
reverse_string = input_lower_string[::-1]

if reverse_string == input_lower_string:
    print("True")
else:
    print("False")
```

```python
"""
Convert a list of tuples to a dictionary
Input: [('a', 1), ('b', 2)]
Output: {'a': 1, 'b': 2}
"""
input_list = [('a', 1), ('b', 2)]
output_dict = {}

def convert_list_to_dict(input_list):
    for key, value in input_list:
        output_dict[key] = value

convert_list_to_dict(input_list)
print(output_dict)
```

```python
"""
Find common elements in two lists
Input: list1 = [1, 2, 3], list2 = [2, 3, 4]
Output: [2, 3]
"""
list1 = [1, 2, 3]
list2 = [2, 3, 4]
common_elements = []

def find_common_elements(list1, list2):
    for i in list1:
        if i in list2:
            common_elements.append(i)

find_common_elements(list1, list2)
print(common_elements)

print(set(list1).intersection(set(list2)))
```

```python
"""
Count word frequencies in a sentence using a dict
Input: "the quick brown fox jumps over the lazy dog"
Output: {'the': 2, 'quick': 1, ..., 'dog': 1}
"""
input_string = "the quick brown fox jumps over the lazy dog"
output_dict = {}

def count_word_sequence(input_string):
	split_list = input_string.split()
	for word in split_list:
		if word in output_dict:
			output_dict[word] += 1
		else:
			output_dict[word] = 1

count_word_sequence(input_string)
print(output_dict)

from collections import Counter
print(Counter(input_string.split()))
```

```python
"""
Count frequencies using dictionary & set
Input: ["cat", "dog", "cat", "bird", "dog"]
â†’ Use set to get unique items and dict to count.
"""
input_list = ["cat", "dog", "cat", "bird", "dog"]
unique_items = set(input_list)
count_dict = {}

for item in unique_items:
    count_dict[item] = input_list.count(item)

print(count_dict)
```

```python
"""
Convert list of dicts into a dict of lists
Input: [{'a': 1, 'b': 2}, {'a': 3, 'b': 4}]
Output: {'a': [1, 3], 'b': [2, 4]}
"""
input_list = [{'a': 1, 'b': 2}, {'a': 3, 'b': 4}]
output_dict = {}

def convert_list_of_dicts(input_list):
	for key in input_list[0].keys():
		output_dict[key] = []
	for item in input_list:
		for key, value in item.items():
			output_dict[key].append(value)

convert_list_of_dicts(input_list)
print(output_dict)
```
