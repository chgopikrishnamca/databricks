```python
%sql
/* Display unique match between teams using self join:

AFG Vs AUS
AFG Vs PAK
AFG Vs IND
IND Vs PAK
AUS Vs PAK
AUS Vs IND
*/

select concat(t1.teams, ' Vs ', t2.teams) from teams as t1 join teams as t2
on t1.teams < t2.teams
```
