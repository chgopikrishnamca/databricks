```python
df = spark.table("sales_data")
df.display()
```

```python
from pyspark.sql.functions import spark_partition_id

# Assign partition ID to each row
df = df.withColumn("partition_id", spark_partition_id())
```

```python
# See row counts per partition
df.groupBy("partition_id").count().orderBy("partition_id").show()
```

```python
df_repart = df.repartition(4)  # choose your desired number of partitions

df_repart_with_pid = df_repart.withColumn("partition_id", spark_partition_id())
df_repart_with_pid.groupBy("partition_id").count().orderBy("partition_id").show()
```
